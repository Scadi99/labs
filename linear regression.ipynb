```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
chirps_rate = [20.0,16.0,19.8,18.4,17.1,15.5,14.7,15.7,15.4,16.3,15.0,17.2,16.0,17.0,14.4]
temps = [88.6,71.6,93.3,84.3,80.6,75.2,69.7,71.6,69.4,83.3,79.6,82.6,80.6,83.5,76.3]
xx = np.linspace(65,95,20)
def lin_model(x,a,b):
    return a*x+b
a,b=curve_fit(lin_model,temps,chirps_rate)[0]
print=(a,b)
plt.plot(temps,chirps_rate,'.')
plt.plot(xx,lin_model(xx,a,b), label='y=%sx+%s'%(round(a,3), round (b,3)))
plt.xlabel('temperature')
plt.ylabel('chirps/sec')
plt.title('change in chirps/sec according to temperature')
plt.legend(loc=4)
plt.ylim([0,21])
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

time = [0,.04,.08,.12, .16,.20,.24]
dis = [0,3.3,5.8,7.7,9.1,9.9,10.2]
xx = np.linspace(0,.50,30)

def quad_model(x,a,b,c):
    return a*x**2+b*x+c

a,b,c= curve_fit(quad_model,time,dis)[0]
print=(a,b,c)                                  

plt.plot(time,dis,'.')
plt.plot(xx,quad_model(xx,a,b,c))
plt.xlabel('time')
plt.ylabel('distance')
plt.title('distance of a bball over time when bounced')
plt.ylim([0,11])
plt.show() 
```
